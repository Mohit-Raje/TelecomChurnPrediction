```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import pickle
```

```python
with open('/content/model1.pkl' , 'rb') as file:
  model=pickle.load(file)

with open('/content/preprocessor1.pkl' , 'rb') as file:
  preprocessor=pickle.load(file)
```

```python
df=pd.read_csv("/content/Final_data.csv")
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0'] , axis=1 , inplace=True)
```

```python
df.head()
```

```python
columns=list(df.columns)
columns
```

```python
print("Enter customer details below:")

tenure = int(input("Tenure (in months): "))
multiple_lines = input("MultipleLines (Yes/No): ")
internet_service = input("InternetService (DSL/Fiber optic/No): ")
online_security = input("OnlineSecurity (Yes/No): ")
tech_support = input("TechSupport (Yes/No): ")
contract = input("Contract (Month-to-month/One year/Two year): ")
paperless_billing = input("PaperlessBilling (Yes/No): ")
payment_method = input("PaymentMethod (Electronic check/Mailed check/Bank transfer (automatic)/Credit card (automatic)): ")
monthly_charges = float(input("MonthlyCharges (e.g., 75.5): "))
total_charges = float(input("TotalCharges (e.g., 820.5): "))

custom_input = {
    'tenure': tenure,
    'MultipleLines': multiple_lines,
    'InternetService': internet_service,
    'OnlineSecurity': online_security,
    'TechSupport': tech_support,
    'Contract': contract,
    'PaperlessBilling': paperless_billing,
    'PaymentMethod': payment_method,
    'MonthlyCharges': monthly_charges,
    'TotalCharges': total_charges
}

custom_df=pd.DataFrame([custom_input])

transformed_data = preprocessor.transform(custom_df)

prediction = model.predict(transformed_data)

print(prediction[0])

if prediction[0]==0:
  print('Retained')
else:
  print('Churn')
```
